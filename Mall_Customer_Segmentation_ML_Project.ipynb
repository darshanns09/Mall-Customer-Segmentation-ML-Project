# K-MEAN AGORITHM

```python
X1=df.loc[:, ["Age", "Spending Score (1-100)"]].values  

'''This line selects two columns, "Age" and "Spending Score (1-100)", from the DataFrame df.
    selects all rows (:) and the specified columns.
    .values = converts the selected columns into a NumPy array (X1), which is a format suitable for use with scikit-learn.'''

from sklearn.cluster import KMeans                        # This imports the KMeans class from scikit-learn's clustering module.
wcss = []

for k in range(1,11):
    kmean = KMeans(n_clusters=k, init="k-means++")
    kmean.fit(X1)
    wcss.append(kmean.inertia_)
    
''' This loop iterates over the range from 1 to 10 (inclusive), representing different numbers of clusters (k).
    For each value of k, it creates a KMeans object with n_clusters=k and init="k-means++".
    The "k-means++" initialization method selects initial cluster centers in a way that speeds up convergence.
    kmeans.fit(X1) = fits the KMeans model to the data X1.
    kmeans.inertia_ = represents the within-cluster sum of squared distances to the nearest centroid. and appends it to the wcss list.'''
    
plt.figure(figsize=(15,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8") # plots the range of cluster numbers (from 1 to 10) against the corresponding WCSS values.
plt.xlabel('K Value')
plt.ylabel('wcss')
plt.show()
```

```python
X1
```

```python
wcss
```

```python
kmean1 = KMeans(n_clusters=4)

label1 = kmean1.fit_predict(X1)

print(label1)
```

```python
print(kmean1.cluster_centers_)
```

```python
plt.scatter(X1[:,0], X1[:,1], c=kmean1.labels_, cmap='rainbow')     #represent the first and second columns of the array X1, respectively. 

plt.scatter(kmean1.cluster_centers_[:,0], kmean1.cluster_centers_[:,1], color = 'black')

plt.title('Clusters of Customer')

plt.ylabel('Spending Score (1-100)')

plt.xlabel('Age')
```
